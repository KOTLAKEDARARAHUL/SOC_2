```python
#to check the mail is spam or ham mail
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
raw_mail_data=pd.read_csv(r"C:\Users\kedar\Downloads\mail_data.csv")
raw_mail_data
```

```python
#to check and Nan filled in column 
raw_mail_data.isna().sum()
```

```python
#to convert category field to numerical field 'spam'=0
raw_mail_data.loc[raw_mail_data['Category']=='ham','Category']=1
raw_mail_data.loc[raw_mail_data['Category']=='spam','Category']=0
raw_mail_data['Category']
```

```python
y=raw_mail_data['Category']
```

```python
x=raw_mail_data['Message']
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#Term frequency Inverse Document Frequency-->It is a numerical representation
#of test document to capture the important words in a collection of document
feature_extraction=TfidfVectorizer()
x_test_feature=feature_extraction.fit_transform(x_test)
print(x_test_feature)
```

```python
#check the type of y_train and y_test
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
y_train.dtype
```

```python
y_test.dtype
```

```python
#Create instance for Logistic Regression
model=LogisticRegression()
model
```

```python
model.fit(x_test_feature,y_test)
```

```python
#process of testing the model
prediction=model.predict(x_test_feature)
prediction
```

```python
acc=accuracy_score(prediction,y_test)
acc
```
